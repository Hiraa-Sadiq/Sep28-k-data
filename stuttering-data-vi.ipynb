```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/fluencybank_labels.csv')
```

Note: A block is gasps for air or stuttered pauses

```python
# Group by 'ID' and sum the other columns
sums = df.groupby(['EpId', 'Block']).size().unstack(fill_value=0)

# Plot the sums for each group
sums.plot(kind='bar', stacked=True)
plt.xlabel('Episode IDs')
plt.ylabel('Sum of Block Column')
plt.title('Sum of Blocks Column by EpId')
plt.legend(loc='upper right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import wave

# Replace 'your_audio.wav' with the path to your WAV file
audio_file = 'output.wav'

# Open the WAV file
wav = wave.open(audio_file, 'r')

# Read audio data and metadata
signal = wav.readframes(-1)
signal = np.frombuffer(signal, dtype='int16')
frame_rate = wav.getframerate()
time = np.linspace(0, len(signal) / frame_rate, num=len(signal))

# Plot the waveform
plt.figure(figsize=(20, 4))
plt.plot(time, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform of the Audio')
plt.grid()
plt.show()

# Close the WAV file
wav.close()
```
